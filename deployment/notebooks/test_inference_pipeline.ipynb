```python
import numpy as np

RNG = np.random.default_rng(10062930)
```

```python
import matplotlib.pyplot as plt

sample = RNG.exponential(scale=100000, size=1)
sample[0]
```

```python
generate_choice = lambda: RNG.choice(
    [
        "PAYMENT",
        "CASH_OUT",
        "CASH_IN",
        "DEBIT",
        "TRANSFER",
    ]
)

generate_amount = lambda: RNG.exponential(scale=100000, size=1)[0]
```

```python
message = {
    "amount": generate_amount(),
    "old_balance_Source": generate_amount(),
    "new_balance_Source": generate_amount(),
    "old_balance_Destination": generate_amount(),
    "new_balance_Destination": generate_amount(),
    "type": generate_choice(),
}
for k, v in message.items():
    if k == "type":
        print(f"{k}: {v}")
    else:
        print(f"{k}: {v:.2f}")
print()

# Convert message JSON into ML model input format.
batch = pd.Series(message).to_frame().T

batch = pi.pipeline(batch)
pred = batch.is_fraud
print(pred[0])
```
